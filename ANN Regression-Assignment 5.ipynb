```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#read dataset
dataset = pd.read_csv("C:\MP/50_Startups.csv")
```

```python
dataset.isnull().any()
```

```python
#to find out null values
dataset.isnull().sum()
```

```python
#fill the null values with the mean value
dataset["Administration"].fillna(dataset["Administration"].mean(),inplace = True)
dataset["Marketing Spend"].fillna(dataset["Marketing Spend"].mean(),inplace = True)
```

```python
dataset.isnull().any()
```

```python
#splitting of input and output data
x = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4:].values
```

```python
x
```

```python
y
```

```python
dataset["State"].unique()
```

```python
#transform the state column into non textual data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("one",OneHotEncoder(),[3])],remainder = "passthrough")
x = ct.fit_transform(x)
```

```python
x.shape
```

```python
y.shape
```

```python
#data splits into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
print('x shape ', x.shape)
print('y shape ', y.shape)
print('x_train shape ', x_train.shape)
print('x_test shape ', x_test.shape)
print('y_train shape ', y_train.shape)
print('y_test shape', y_test.shape)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
regression = Sequential()
```

```python
#adding input layer
regression.add(Dense(units =6 ,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
#adding hidden layer
regression.add(Dense(units =12 ,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
regression.add(Dense(units =12 ,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
#adding output layer
regression.add(Dense( units = 1 , kernel_initializer="random_uniform" , activation = "relu")) 
```

```python
regression.compile(optimizer = "adam",loss = "mse",metrics = ["mse"]) 
```

```python
#train the model
regression.fit(x_train,y_train , epochs = 20000)
```

```python
ypred=regression.predict(x_test)
```
